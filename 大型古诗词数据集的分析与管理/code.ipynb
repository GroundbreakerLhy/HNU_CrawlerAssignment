**2.显示超过200首诗歌的唐代诗人并对诗人的诗词数量进行饼图可视化**

```python
from collections import Counter
import matplotlib.pyplot as plt
# 支持中文
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号

poets_counter = Counter(item[3] for item in data.values) # 使用Counter类直接统计数量

poets_with20_poems = {poet: count for poet, count in poets_counter.items() if count > 200} # 筛选出超过200首诗的诗人

print(poets_with20_poems)

plt.figure(figsize=(20, 20))
font_properties = {'fontsize': 24} # 设置图片尺寸

def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        absolute = int(round(pct * total / 100.0))
        return f"{pct:.1f}%\n({absolute:d})"
    return my_format

plt.pie(poets_with20_poems.values(), labels=poets_with20_poems.keys(), autopct=autopct_format(poets_with20_poems.values()))

plt.title('超过200首诗歌的唐代诗人')

plt.show()
```
